```python
# ================================
# IMPORT REQUIRED LIBRARIES
# ================================

# Prompt templates for structured input
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

# OpenAI LLM wrapper (if needed later)
from langchain_openai import OpenAI

# Chain for combining prompt + model
from langchain.chains import LLMChain

# Memory to store conversation history
from langchain.memory import ConversationBufferMemory

# Agent initialization (not used yet, but part of LangChain ecosystem)
from langchain.agents import initialize_agent

# Tool abstraction (used in agents)
from langchain.tools import Tool


# ================================
# PROMPT TEMPLATE EXAMPLE
# ================================

print("\n🔹 Prompt Template Example:")

# Create a reusable prompt template
template = PromptTemplate(
    input_variables=["topic"],  # variable input
    template="Explain {topic} in simple terms"  # template structure
)

# Format the prompt with actual input
formatted_prompt = template.format(topic="LangChain")

print(formatted_prompt)


# ================================
# MEMORY EXAMPLE
# ================================

print("\n🔹 Memory Example:")

# Initialize conversation memory
memory = ConversationBufferMemory()

# Save conversation context (user input + AI response)
memory.save_context(
    {"input": "Hello"},
    {"output": "Hi there!"}
)

memory.save_context(
    {"input": "What is AI?"},
    {"output": "AI stands for Artificial Intelligence"}
)

# Load stored memory
print(memory.load_memory_variables({}))


# ================================
# CHAT PROMPT TEMPLATE EXAMPLE
# ================================

print("\n🔹 Chat Prompt Example:")

# Create a chat-style prompt (system + user roles)
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),  # system instruction
    ("user", "Explain {topic}")  # user input
])

# Format chat prompt
print(chat_template.format(topic="Neural Networks"))


# ================================
# INPUT VALIDATION FUNCTION
# ================================

print("\n🔹 Input Validation:")

# Function to validate user inputs
def validate_inputs(audience, tone):
    # Allowed values
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Default fallback if invalid
    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Test validation
print(validate_inputs("kids", "cool"))


# ================================
# PROMPT GENERATOR FUNCTION
# ================================

print("\n🔹 Prompt Generator:")

# Create advanced prompt template
teaching_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} teaching style"
)

# Function to generate structured prompt
def generate_prompt(topic, audience, tone):
    # Validate inputs first
    audience, tone = validate_inputs(audience, tone)

    # Return formatted prompt
    return teaching_prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test prompt generator
print(generate_prompt("Machine Learning", "beginner", "fun"))


# ================================
# TEMPLATE REUSABILITY TEST
# ================================

print("\n🔹 Template Reusability Test:")

# Reusable prompt template
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple words"
)

# Multiple topics to demonstrate reuse
topics = ["AI", "Python", "ML", "DL", "Data Science"]

# Loop through topics and generate prompts
for t in topics:
    print(reuse_template.format(topic=t))
```
